```python
import pandas as pd
from pathlib import Path
```

```python
data_path = Path ("../../../projects/web_scraping_analysis/data/jobs.csv")
df = pd.read_csv(data_path)
df.head()
```

### Top Job Titles

```python
top_titles = df["title"].value_counts().head(10)
top_titles
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
top_titles.sort_values().plot(kind="barh")
plt.title("Top 10 Job Titles")
plt.xlabel("Count")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

## Job Title Insight

This analysis shows the most common job titles in the scraped dataset and helps identify which roles appear most frequently.
