```python
import pandas as pd
from pathlib import Path
```

```python
data_path = Path ("../../../projects/web_scraping_analysis/data/jobs.csv")
df = pd.read_csv(data_path)
df.head()
```

```python
df.shape
```

```python
df.columns
```

### Top Job Titles

```python
top_titles = df["title"].value_counts().head(10)
top_titles
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
top_titles.sort_values().plot(kind="barh")
plt.title("Top 10 Job Titles")
plt.xlabel("Count")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

## Job Title Insight

This analysis shows the most common job titles in the scraped dataset and helps identify which roles appear most frequently.

### Top Companies

```python
top_companies = df["company"].value_counts().head(10)
top_companies
```

```python
plt.figure(figsize=(10, 6))
top_companies.sort_values().plot(kind="barh")
plt.title("Top 10 Companies Hiring")
plt.xlabel("Number of Listings")
plt.ylabel("Company")
plt.tight_layout()
plt.show()
```

## Company Insight

This analysis highlights which companies appear most frequently in the dataset and helps identify major hiring organizations.

### Top Locations

```python
top_locations = df["location"].value_counts().head(10)
top_locations
```

```python
plt.figure(figsize=(10, 6))
top_locations.sort_values().plot(kind="barh")
plt.title("Top 10 Job Locations")
plt.xlabel("Number of Listiings")
plt.ylabel("Location")
plt.tight_layout()
plt.show()
```

## Location Insight

This analysis shows where job opportunities are concentrated and helps identify the most common locations in the dataset.
