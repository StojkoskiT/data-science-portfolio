```python
import pandas as pd 
from pathlib import Path
```

```python
data_path = Path("../data/sentiment_data.csv")
df = pd.read_csv(data_path)

df.head()
```

```python
df["sentiment"].value_counts()
```

### Cleaning and preparing the text

```python
df["text"] = df["text"].str.lower()
df["text"]
```

```python
import re

df["clean_text"] = df["text"].apply(lambda x: re.sub(r"[^a-zA-z\s]", "", x))
df[["text", "clean_text"]]
```

## Text Preprocessing

The text was converted to lowercase and cleaned by removing punctuation and non-letter characters.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["clean_text"])
```

```python
y = df["sentiment"].map({"negative": 0, "positive": 1})
y
```
